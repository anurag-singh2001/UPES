```python
import scipy.stats as stats
```

```python
x = np.linspace(1,50,20)
x_data=pd.DataFrame(x)

z=stats.zscore(x_data)
z_data=pd.DataFrame(z)

print("Z DISTRIBUTION")
print(z_data)
```

```python
x = np.linspace(1,50,20)
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
mean = np.mean(x)
sd = np.std(x)
pdf = normal_dist(x,mean,sd)
plt.plot(x,pdf , color = 'blue')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
plt.title("PLOTTING NORMAL DISTRIBUTION")
plt.show()
```

```python
x = np.linspace(1,50,20)
z=stats.zscore(x)
def normal_dist(z , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((z-mean)/sd)**2)
    return prob_density
mean = np.mean(z)
sd = np.std(z)
pdf = normal_dist(z,mean,sd)
plt.plot(z,pdf, color = 'red')
plt.xlabel('Z Data points')
plt.ylabel('Z Probability Density')
plt.title("PLOTTING Z DISTRIBUTION")
plt.show()
```
