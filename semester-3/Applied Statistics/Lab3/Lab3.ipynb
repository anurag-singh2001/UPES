```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mu, sigma = 5, 2.2 
s = np.random.normal(mu, sigma, 10)
s
```

```python
s.mean()
```

```python
s.std()
```

```python
df = pd.DataFrame({
      'values': s,
      'mean': [5 for i in range(len(s))],
      'std': [2.2 for i in range(len(s))]})

df.plot()
plt.show()
```
