```python
import pandas as pd
```

```python
df=pd.read_csv('dataset.csv',encoding = "ISO-8859-1")
df
```

```python
y=df['Label']
y
```

```python
x= df.drop(['Date', 'Label'], axis=1)
x
```

```python
x['combined'] = x.apply(lambda y: ''.join(y.astype(str)), axis=1)
x['combined']
```

```python
import string

def remove_non_ascii(s):
    """
    Remove non-ASCII characters from a string.
    """
    printable = set(string.printable)
    return ''.join(filter(lambda x: x in printable, s))



# apply the remove_non_ascii function to each element in the 'text' column
x['text'] = x['combined'].apply(lambda x: remove_non_ascii(x))

# display the updated DataFrame
print(x['text'])
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
x['text'] = x['text'].apply(lambda i: i.lower() if isinstance(i, str) else i)
x['text']
```

```python
import nltk
from nltk.tokenize import sent_tokenize
x['text'] = x['text'].apply(lambda x: sent_tokenize(x))
print(x['text'])
```

```python
from nltk.tokenize import word_tokenize
x['text'] = x['text'].astype(str)
x['words'] = x['text'].apply(lambda x: word_tokenize(x))
print(x['text'])
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
x['stemmed_words'] = x['words'].apply(lambda x: [stemmer.stem(word) for word in x])
print(x['stemmed_words'])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
x['lemmatized_words'] = x['words'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
print(x['lemmatized_words'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
x['lemmatized_text'] = x['lemmatized_words'].apply(lambda x: " ".join(x))
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(x['lemmatized_text']).toarray()
print(bow)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

x['lemmatized_text'] = x['lemmatized_words'].apply(lambda x: " ".join(x))
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(x['lemmatized_text']).toarray()
print(tfidf)
```

```python
# import required libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords

# concatenate all reviews into a single string
reviews = x['lemmatized_words'].apply(lambda x: ' '.join(x))
text = ' '.join(reviews)

# remove stop words from the text
stopwords = set(stopwords.words("english"))
text = ' '.join([word for word in text.split() if word not in stopwords])
# stopwords = set(stopwords.words("hindi"))
# text = ' '.join([word for word in text.split() if word not in stopwords])

# create a wordcloud object
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(text)

# plot the wordcloud
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show()
```
