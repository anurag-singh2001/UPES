```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
df['renderedContent'] = df.renderedContent.str.lower()
print(df['renderedContent'])
```

```python
import nltk
from nltk.tokenize import sent_tokenize
df['sentences'] = df['renderedContent'].apply(lambda x: sent_tokenize(x))
print(df['sentences'])
```

```python
from nltk.tokenize import word_tokenize

df['words'] = df['renderedContent'].apply(lambda x: word_tokenize(x))

print(df['words'])
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
df['stemmed_words'] = df['words'].apply(lambda x: [stemmer.stem(word) for word in x])
print(df['stemmed_words'])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
df['lemmatized_words'] = df['words'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
print(df['lemmatized_words'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
df['lemmatized_text'] = df['lemmatized_words'].apply(lambda x: " ".join(x))
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(df['lemmatized_text']).toarray()
print(bow)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

df['lemmatized_text'] = df['lemmatized_words'].apply(lambda x: " ".join(x))
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(df['lemmatized_text']).toarray()
print(tfidf)
```

```python
# import required libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords

# concatenate all reviews into a single string
reviews = df['lemmatized_words'].apply(lambda x: ' '.join(x))
text = ' '.join(reviews)

# remove stop words from the text
stopwords = set(stopwords.words("english"))
text = ' '.join([word for word in text.split() if word not in stopwords])
# stopwords = set(stopwords.words("hindi"))
# text = ' '.join([word for word in text.split() if word not in stopwords])

# create a wordcloud object
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(text)

# plot the wordcloud
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show()
```

```python
import gensim

model = gensim.models.Word2Vec(df['lemmatized_words'], size=100, window=5, min_count=1, workers=4)
word_vectors = model.wv
```

```python
embeddings_df = pd.DataFrame(columns=["word", "embedding"])
for word in word_vectors.vocab:
  embeddings_df = embeddings_df.append({"word": word, "embedding": word_vectors[word]}, ignore_index=True)
embeddings_df.to_csv("embeddings.csv", index=False)
```
